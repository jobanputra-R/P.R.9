```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
class DataLoading:
    def loader (self):
        self.name=input("enter name of file to load")
        self.data=pd.read_csv(self.name)
```

```python
class Explore(DataLoading):
    def explore (self):
     print(" == Explore Data == ")
     print("1. Display the first 5 rows")
     print("2. Display the last 5 rows")
     print("3. Display column names")
     print("4. Display data types")
     print("5. Display basic info")

     data=self.data  
     choice = input("Enter Your Choice:- ")
     if choice =="1":
         print(data.head(5))
     elif choice == "2":
          print(data.tail(5))
     elif choice == "3":
         for col in data.columns:
             print(col)
     elif choice == "4":
         print(data.dtypes)   
     elif choice == "5":
          print(data.info())  
     else:
         print("Invalid Option !")

#obj=Explore()
#obj.loader()
#obj.explore()
```

```python
class DataFrame_Operations(Explore):
  def df_operations(self):
     print("1. Filter Rows")
     print("2. Sort Data")
     print("3. Show column name ")
     print("4. Group By")
     ch=input("Enter Your Choice!")
     if ch=="1":
        col=input("Enter Column name: ")
        val=float(input("Enter value: "))
        print(self.data[self.data[col]>val])
     elif ch=="2":
        col=input("Enter Column to sort by: ")
        print(self.data.sort_values(by=col))
     elif ch=="3":
        print(self.data.columns)
     elif ch=="4":
        col=input("1.Enter group by colum name: ")
        col2=input("2.Enter group by colum name: ")
        print(self.data.groupby(col)[col2].sum())
     else:
        print("Wrong Input !")
```

```python
class MissingData(DataFrame_Operations):
   def missing(self):
     print(" == Handel Missing values == ")
     print("1. Display rows with missing values")
     print("2. Fill missing values with mean")
     print("3. Drop rows with missing values")
     print("4. Replace missing values with a specific value")
     ch=input("Enter Your Choice!")
     if ch=="1":
       missing=self.data[self.data.isna().any(axis=1)]
       print(missing.to_string())
     elif ch=="2":
       filled=self.data.fillna(self.data.mean(numeric_only=True))
       print("Data filled with mean")
       print(filled)
     elif ch=="3":
       droped=self.data.dropna()
       print("Rows dropped !")
       print(droped)
     elif ch=="4":
       fill_with=self.data.fillna(input("Enter values to fill"))
       print("Values updated  with your value") 
       print(fill_with)
     else :
        print("Invalid Option !")
```

```python
class Stats(MissingData):
     def stats(self):
      print("Generating Descriptive Statistics")
      print("1. whole data frame stats")
      print("2. specific columns")
      choice=(input("Enter your choice"))
      if choice=="1":
         print(self.data.describe())
      elif choice=="2":
         print(self.data[input("Enter column name")].describe())
      else:
         print("Wrong Input !")
```

```python
class DataVisualization(Stats):
    def visual(self):
     print("== Data Visualization ==")
     print("1. Bar Plot")
     print("2. Line Plot")
     print("3. Scatter Plot")
     print("4. Pie Chart")
     print("5. Histogram")
     print("6. Stack Plot")
     choice=(input("Enter Your Choice"))
     if choice=="1":
            title=input("Enter title of Bar Plot")
            x_axis=input("enter X-axis data column name")
            y_axis=input("enter Y-axis data column name")
            plt.figure(figsize=(10,8))
            sns.barplot(x=x_axis,y=y_axis,data=self.data.dropna())
            plt.xlabel(x_axis)
            plt.ylabel(y_axis)
            plt.title(title)
            plt.tight_layout()
            plt.show()
     elif choice=="2":
            title=input("Enter title of Line Plot")
            x_axis=input("enter X-axis data column name")
            y_axis=input("enter Y-axis data column name")
            plt.figure(figsize=(12,10))
            sns.lineplot(x=x_axis,y=y_axis,data=self.data.dropna())
            plt.xlabel(x_axis)
            plt.ylabel(y_axis)
            plt.title(title)
            plt.tight_layout()
            plt.show()
     elif choice=="3":
            title=input("Enter title of Scatter Plot")
            x_axis=input("enter X-axis data column name")
            y_axis=input("enter Y-axis data column name")
            plt.figure(figsize=(12,10))
            sns.scatterplot(x=x_axis,y=y_axis,data=self.data.dropna())
            plt.xlabel(x_axis)
            plt.ylabel(y_axis)
            plt.title(title)
            plt.tight_layout()
            plt.show()
     elif choice=="4":
            title=input("Enter title of Pie Chart")
            y_axis=input("Enter column name for pie chart")
            data = self.data[y_axis].value_counts()
            top_n = int(input("Enter number of category to show (eg.5)"))
            data = data.head(top_n)
            sns.set_style("whitegrid")
            plt.figure(figsize=(12,10))
            plt.pie(data.values,labels=data.index,autopct="%0.2f%%",startangle=90)
            plt.ylabel(y_axis)
            plt.title(title)
            plt.tight_layout()
            plt.show()
     elif choice=="5":
            title=input("Enter Title of Histogram")
            y_axis=input("Enter column name for Histogram")
            data = self.data[y_axis].value_counts()
            top_n = int(input("Enter number of category to show (eg.5)"))
            data = data.head(top_n)
            plt.figure(figsize=(12,10))
            sns.histplot(data.dropna(),bins=5,kde=True)
            plt.title(title)
            plt.tight_layout()
            plt.show()
     elif choice=="6":
            title=input("Enter title of Scatter Plot")
            x_axis=input("enter X-axis data column name")
            y_axis=input("enter Y-axis data column name")
            plt.figure(figsize=(12,10))
            plt.stackplot(x=x_axis,y=y_axis,data=self.data.dropna())
            plt.xlabel(x_axis)
            plt.ylabel(y_axis)
            plt.title(title)
            plt.tight_layout()
            plt.show()
     else:
            print("Wrong Input!")
```

```python
class Save_visual(DataVisualization):
    def save_visualization(self):
        print("== Save Visualization")
        fname=input("Enter file name to save the plot (e.g., Scatter_plot.png): ")
        print("Choose format ")
        print("1.PNG")
        print("2.JPG")
        print("3.PDF")
        choice=input("Enter choice")
        if choice =="1":
            ext="png"
        elif choice =="2":
            ext="jpg"
        elif choice =="3":
            ext="pdf"
        else:
            print("Invalid Input")
        fullname=f"{fname}.{ext}"
        plt.savefig(fullname,bbox_inches="tight")  
        print(f"Visulization saved as {fullname} successfully!") 
obj=Save_visual()
 
```

```python
while True:
    print("========== Data Analysis & Visualization Program ==========")
    print("1. Load Dataset")
    print("2. Explore Data")
    print("3. Perform DataFrame Operations")
    print("4. Handle Missing Data")
    print("5. Generate Descriptive Statistics")
    print("6. Data Visualization")
    print("7. Save Visualization")
    print("8. Exit")
    print("="*59)
    Choice=int(input("Enter your choice :"))
    match Choice:
        case 1:
            obj.loader()
        case 2:
            obj.explore()
        case 3:
            obj.df_operations()
        case 4:
            obj.missing()
        case 5:
            obj.stats()
        case 6:
            obj.visual()
        case 7:
            obj.save_visualization()
        case 8:
            print("Exiting the program. Goodbye!")
            break
        case _:
            print("OOPS! Wrong Option Selected")
            continue
```
